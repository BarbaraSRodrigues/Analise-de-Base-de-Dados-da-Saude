## Escolha da base de dados

As doenças cardíacas são uma das principais causas de morte em todo o mundo. Elas afetam uma grande parte da população e têm um impacto significativo na saúde pública.

Segundo o site Cardiômetro, um indicador do número de mortes por doenças cardiovasculares no País, criado pela Sociedade Brasileira de Cardiologia, as doenças cardiovasculares, afecções do coração e da circulação, representam a principal causa de mortes no Brasil. São mais de 1100 mortes por dia, cerca de 46 por hora, além disso, essas doenças matam 2 vezes mais que todos os tipos de câncer somados. Além disso, segundo o G1, atualmente, cerca de 14 milhões de pessoas têm alguma enfermidade cardiovascular no Brasil. Apesar da importância da prevenção, 23% dos brasileiros nunca foram ao cardiologista.

Dessa forma, a preocupação com doenças cardíacas é justificada pela prevalência, gravidade e impacto que esse tipo de doença têm na saúde e qualidade de vida da população. Nesse sentido, é importante adotar medidas preventivas e buscar cuidados médicos adequados para mitigar os riscos associados a doenças cardíacas. Portanto, escolhemos essa base de dados devido a relevância desse assunto nos dias atuais, principalmente na área da saúde.

## Mapa de Calor e Correlação

Um mapa de calor é uma representação gráfica dos dados, que destaca a densidade dos dados de acordo com uma rubrica codificada por cores ou outra forma de chave de mapeamento.

```python
# Tabela de correcao entre os atributos de cada instancia para o uso no mapa de calor
corr = data.corr()
corr
```

```python
# Configurando o tamando da figura
f, ax = plt.subplots(figsize=(15, 15))
plt.title('Mapa de calor do dataset')

# Utilizando a biblioteca Seaborn para visualizar o mapa de calor
sns.heatmap(data.astype(float).corr(),linewidths=0.7,vmax=1.0, square=True, annot=True, cmap='inferno')
```

O mapa de calor mostra que as correlações variam de **-1 a 1**, onde quanto mais próximo de 1, mais uma variável possui uma **correlação positiva forte** com a outra, ou seja, conforme a variável x aumenta, a variável y tende a aumentar, assim como quando a variável x diminui, a variável y também tende a diminuir, e vice e versa. Podemos ter uma **correlação negativa forte** onde temos uma valor de correlação próximo a -1 (tende a aumentar, a variável y tende a diminuir, e vice e versa)

## Relações entre atributos (de pares)

Com a plotagem desse gráfico, podemos analisar melhor a relação entre as variáveis do dataset

```python
#Plotagem em pares dos atributos da base de dados.
sns.pairplot(data = data, vars=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target'])
plt.show()
```

## Correlação de Pearson

Coeficiente que mede a correlação linear entre duas variáveis. Seu valor varia entre -1 e 1, onde 1 indica uma correlação linear positiva perfeita, -1 indica uma correlação linear negativa perfeita e 0 indica nenhuma correlação.

image.png

```python
correlations = data.corr()

target_correlations = correlations['target'].drop('target')  # Exclui a propria var

# Plotando o grafico de barras horizontalmente
plt.figure(figsize=(10, 6))
plt.barh(target_correlations.index, target_correlations.values)
plt.title('Correlação de Pearson com a variável target')
plt.xlabel('Coeficiente de Correlação de Pearson')
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

```python
#Calculando o coeificiente de correlação entre algumas variáveis da base de dados;

data[['chol', 'trestbps', 'fbs', 'restecg', 'thalach', 'thal']].corr()
```

```python
#Apresentando o gráfico PairGrid em relação a algumas variáveis
#da base de dados.

g = sns.PairGrid(data[['chol', 'trestbps', 'fbs', 'restecg', 'thalach', 'thal']])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

Como não foi achada correlacão fortemente positiva ou negativa, podemos observar pelos gráficos que eles não apresentam correlações fortes.

```python
#Apresentando gráficos BoxPlot de algumas variáveis em relação a mpg.

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(19,6))

fig.suptitle("Gráfico BoxPlot para algumas variáveis x idade", fontsize=20)

sns.boxplot(x="age", y="chol", data=data,ax=ax1)#Colesterol
sns.boxplot(x="age", y="trestbps", data=data,ax=ax2)#Pressão Arterial
```

No BoxPlot, podemos analisar que há alguns outliers, ou seja, alguns valores fogem da média e tem valores atípicos.
